# Importación de librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Configuración para visualizaciones
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
```

# Preprocesamiento de los datos

Esta sección detalla el proceso de carga, combinación y limpieza de los datos de nacimientos. Es fundamental preparar los datos adecuadamente para asegurar la calidad y fiabilidad del modelo de Machine Learning.

```python
# Rutas de los archivos CSV
file_paths = {
    '2020': './data/nac2020.csv',
    '2021': './data/nac2021.csv',
    '2022': './data/nac2022.csv',
    '2023': './data/BD-EEVV-Nacimientos-2023.csv'
}

# Lista para almacenar los DataFrames individuales
dfs = []

# Cargar cada archivo CSV
for year, path in file_paths.items():
    try:
        # Intentar leer con codificación 'latin1' (común en archivos de datos de Latinoamérica)
        df = pd.read_csv(path, encoding='latin1', low_memory=False)
        dfs.append(df)
        print(f"Cargado {path} con {len(df)} filas.")
    except FileNotFoundError:
        print(f"Error: El archivo {path} no fue encontrado. Asegúrate de que está en la misma carpeta.")
    except UnicodeDecodeError:
        print(f"Error de decodificación Unicode para {path}. Intentando con otra codificación o revisa el archivo.")
        # Si 'latin1' falla, podrías intentar 'ISO-8859-1' o 'cp1252'
        # Por simplicidad, aquí solo intentamos 'latin1' como la solución más común.
    except Exception as e:
        print(f"Error al cargar {path}: {e}")

# Concatenar todos los DataFrames en uno solo
if dfs:
    df_combined = pd.concat(dfs, ignore_index=True)
    print(f"\nTotal de filas combinadas: {len(df_combined)}")
else:
    df_combined = pd.DataFrame()
    print("No se pudo cargar ningún archivo CSV.")

# Mostrar las primeras filas del DataFrame combinado
print("\nPrimeras 5 filas del DataFrame combinado:")
df_combined.head()
```

# Manejo de valores faltantes

Se identificarán las columnas con valores nulos y se aplicará una estrategia de imputación o eliminación según la relevancia de la columna para el modelo y la cantidad de datos faltantes.

```python
# Información general del DataFrame y conteo de valores nulos
print("\nInformación del DataFrame combinado:")
df_combined.info()

print("\nValores nulos por columna:")
missing_values = df_combined.isnull().sum()
print(missing_values[missing_values > 0].sort_values(ascending=False))

# Estrategia de manejo de nulos:
# 1. 'OTRO_SIT': Tiene muchos nulos y es una descripción específica. No es directamente relevante para la tasa de natalidad general, así que se puede eliminar.
# 2. 'FECHA_NACM': Fecha de nacimiento de la madre. Muchos nulos. Para la predicción de nacimientos por mes/año, no es una variable clave. Se puede eliminar.

# 'OTRO_SIT' y 'FECHA_NACM': Se eliminan por tener muchos nulos y no ser directamente relevantes para la predicción agregada.
columns_to_drop = ['OTRO_SIT', 'FECHA_NACM']
df_processed = df_combined.drop(columns=columns_to_drop, errors='ignore')

print(f"\nColumnas eliminadas: {columns_to_drop}")
print(f"Filas restantes después de eliminar columnas: {len(df_processed)}")

# Estandarización de valores especiales (9 y 99) que indican 'Sin información'
# Se aplicará a columnas numéricas que representan categorías o códigos.
# Se excluyen 'ANO', 'MES', 'PESO_NAC', 'TALLA_NAC', 'EDAD_MADRE', 'EDAD_PADRE',
# ya que 9/99 no tendría sentido como valor faltante en ellas o son medidas directas.
cols_to_standardize = [
    col for col in df_processed.columns
    if df_processed[col].dtype in ['int64', 'float64'] and
    col not in ['ANO', 'MES', 'PESO_NAC', 'TALLA_NAC', 'EDAD_MADRE', 'EDAD_PADRE']
]

for col in cols_to_standardize:
    # Reemplazar 9 y 99 con NaN en las columnas identificadas
    df_processed[col] = df_processed[col].replace({9: np.nan, 99: np.nan})
    # Si la columna era originalmente int, convertirla a float después de introducir NaN
    if df_combined[col].dtype == 'int64': # Usar df_combined para el dtype original
        df_processed[col] = df_processed[col].astype(float)

print("\nValores nulos después de estandarizar 9 y 99:")
missing_values_after_standardization = df_processed.isnull().sum()
print(missing_values_after_standardization[missing_values_after_standardization > 0].sort_values(ascending=False))


# Para el análisis departamental, es crucial que COD_DPTO no tenga nulos.
# Si después de la estandarización hay nulos en COD_DPTO, se eliminarán esas filas.
if df_processed['COD_DPTO'].isnull().any():
    initial_rows = len(df_processed)
    df_processed.dropna(subset=['COD_DPTO'], inplace=True)
    print(f"\nSe eliminaron {initial_rows - len(df_processed)} filas con COD_DPTO nulo.")
    print(f"Filas restantes después de eliminar COD_DPTO nulos: {len(df_processed)}")

# Asegurarse de que COD_DPTO es de tipo entero para la agrupación
df_processed['COD_DPTO'] = df_processed['COD_DPTO'].astype(int)


# Verificar duplicados
print("\nNúmero de filas duplicadas ANTES de eliminar:", df_processed.duplicated().sum())
df_processed.drop_duplicates(inplace=True)
print("Filas después de eliminar duplicados:", len(df_processed))
```

# Agregación de datos y creación de series temporales

Para ambos análisis, nacional y departamental, se agregará el número de nacimientos por mes y año. Esto nos permitirá crear dos series temporales distintas, una para el total nacional y otra para el total por departamento. También se creará una columna de fecha para facilitar el análisis temporal.

```python
# Agrupar por año y mes para obtener el total de nacimientos a nivel nacional
monthly_births_national = df_processed.groupby(['ANO', 'MES']).size().reset_index(name='total_births')

# Crear una columna de fecha para la serie temporal nacional
monthly_births_national['date'] = pd.to_datetime(
    monthly_births_national['ANO'].astype(str) + '-' +
    monthly_births_national['MES'].astype(str) + '-01'
)

# Ordenar por fecha
monthly_births_national = monthly_births_national.sort_values('date').reset_index(drop=True)

print("\nSerie temporal de nacimientos mensuales (Nacional - primeras filas):")
monthly_births_national.head()
```

```python
# Agrupar por departamento, año y mes para obtener el total de nacimientos
monthly_births_departmental = df_processed.groupby(['COD_DPTO', 'ANO', 'MES']).size().reset_index(name='total_births')

# Crear una columna de fecha para la serie temporal departamental
monthly_births_departmental['date'] = pd.to_datetime(
    monthly_births_departmental['ANO'].astype(str) + '-' +
    monthly_births_departmental['MES'].astype(str) + '-01'
)

# Ordenar por departamento y luego por fecha
monthly_births_departmental = monthly_births_departmental.sort_values(['COD_DPTO', 'date']).reset_index(drop=True)

print("\nSerie temporal de nacimientos mensuales por departamento (primeras filas):")
monthly_births_departmental.head()
```

Se añadirán características para capturar la estacionalidad (mes) y la tendencia (año) que son importantes para la predicción de series temporales. Para el modelo departamental, se incluirá el one-hot encoding del código del departamento.

```python
# Crear características cíclicas para el mes para capturar la estacionalidad (Nacional)
monthly_births_national['month_sin'] = np.sin(2 * np.pi * monthly_births_national['MES']/12)
monthly_births_national['month_cos'] = np.cos(2 * np.pi * monthly_births_national['MES']/12)

# Usar el año como una característica numérica (Nacional)
monthly_births_national['year_numeric'] = monthly_births_national['ANO']

print("\nSerie temporal nacional con características adicionales (primeras filas):")
monthly_births_national.head()
```

```python
# Crear características cíclicas para el mes para capturar la estacionalidad (Departamental)
monthly_births_departmental['month_sin'] = np.sin(2 * np.pi * monthly_births_departmental['MES']/12)
monthly_births_departmental['month_cos'] = np.cos(2 * np.pi * monthly_births_departmental['MES']/12)

# Usar el año como una característica numérica (Departamental)
monthly_births_departmental['year_numeric'] = monthly_births_departmental['ANO']

# Convertir COD_DPTO a tipo 'category' para facilitar el one-hot encoding
monthly_births_departmental['COD_DPTO'] = monthly_births_departmental['COD_DPTO'].astype('category')

# Aplicar One-Hot Encoding a COD_DPTO
# Esto creará una nueva columna binaria para cada departamento
df_features_departmental = pd.get_dummies(monthly_births_departmental, columns=['COD_DPTO'], prefix='DPTO')

print("\nDataFrame departamental con características adicionales y One-Hot Encoding (primeras filas):")
df_features_departmental.head()
```

# Selección del modelo
Para ambos problemas de predicción (nacional y departamental), se ha elegido el algoritmo RandomForestRegressor (Regresor de Bosque Aleatorio). La justificación para esta elección se basa en las siguientes consideraciones, aplicables a ambos niveles de granularidad:

- Naturaleza de los datos: Aunque los datos tienen un componente temporal, la predicción se puede abordar como un problema de regresión donde las características de entrada son el año y el mes (transformado para capturar la estacionalidad). Los Random Forests son muy efectivos en problemas de regresión con datos tabulares.
- Capacidad para capturar relaciones no lineales: La tasa de natalidad puede estar influenciada por factores complejos y no lineales a lo largo del tiempo (ej. cambios socioeconómicos, políticas). Los Random Forests, al ser un ensamble de árboles de decisión, son capaces de modelar relaciones no lineales y capturar interacciones complejas entre las características sin necesidad de una especificación explícita.
- Robustez a outliers y datos ruidosos: Los árboles de decisión son relativamente robustos a los valores atípicos (outliers) y al ruido en los datos, ya que las divisiones se basan en umbrales y no en la magnitud exacta de los valores. Esto es beneficioso dado que los datos demográficos pueden presentar variaciones.
- Manejo de Características Cíclicas: Al transformar el mes en características seno y coseno, el Random Forest puede interpretar mejor la estacionalidad cíclica sin asumir una relación lineal.
- Menor riesgo de sobreajuste (Overfitting): Aunque un solo árbol de decisión puede sobreajustarse, el Random Forest mitiga este riesgo al promediar las predicciones de múltiples árboles entrenados en diferentes subconjuntos de datos (bagging) y características (feature randomness). Esto resulta en un modelo más generalizable.
- Facilidad de uso y buen rendimiento general: Los Random Forests son relativamente fáciles de implementar y suelen ofrecer un buen rendimiento en una amplia gama de problemas de regresión, sirviendo como una excelente línea base.

```python
# Definir las características (X) y la variable objetivo (y) para el modelo NACIONAL
features_national = ['year_numeric', 'month_sin', 'month_cos']
target_national = 'total_births'

X_national = monthly_births_national[features_national]
y_national = monthly_births_national[target_national]

# Dividir los datos en conjuntos de entrenamiento y prueba (Nacional)
# Se utilizará una división basada en el tiempo: los años anteriores para entrenar, el último año para probar.
train_data_national = monthly_births_national[monthly_births_national['ANO'] < 2023]
test_data_national = monthly_births_national[monthly_births_national['ANO'] == 2023]

X_train_national = train_data_national[features_national]
y_train_national = train_data_national[target_national]

X_test_national = test_data_national[features_national]
y_test_national = test_data_national[target_national]

print(f"Tamaño del conjunto de entrenamiento (Nacional): {len(X_train_national)} muestras")
print(f"Tamaño del conjunto de prueba (Nacional): {len(X_test_national)} muestras")
```

```python
# Definir las características (X) y la variable objetivo (y) para el modelo DEPARTAMENTAL
features_departmental = ['year_numeric', 'month_sin', 'month_cos'] + [col for col in df_features_departmental.columns if 'DPTO_' in col]
target_departmental = 'total_births'

X_departmental = df_features_departmental[features_departmental]
y_departmental = df_features_departmental[target_departmental]

# Dividir los datos en conjuntos de entrenamiento y prueba (Departamental)
# Se utilizará una división basada en el tiempo: los años anteriores para entrenar, el último año para probar.
train_data_departmental = df_features_departmental[df_features_departmental['ANO'] < 2023]
test_data_departmental = df_features_departmental[df_features_departmental['ANO'] == 2023]

X_train_departmental = train_data_departmental[features_departmental]
y_train_departmental = train_data_departmental[target_departmental]

X_test_departmental = test_data_departmental[features_departmental]
y_test_departmental = test_data_departmental[target_departmental]

print(f"Tamaño del conjunto de entrenamiento (Departamental): {len(X_train_departmental)} muestras")
print(f"Tamaño del conjunto de prueba (Departamental): {len(X_test_departmental)} muestras")

# Verificar que los departamentos en train y test son consistentes
train_departments = [col.replace('DPTO_', '') for col in X_train_departmental.columns if 'DPTO_' in col]
test_departments = [col.replace('DPTO_', '') for col in X_test_departmental.columns if 'DPTO_' in col]

if set(train_departments) != set(test_departments):
    print("\nAdvertencia: Los departamentos presentes en el conjunto de entrenamiento y prueba difieren.")
    print("Departamentos en entrenamiento:", sorted(list(set(train_departments))))
    print("Departamentos en prueba:", sorted(list(set(test_departments))))
    # Esto puede ocurrir si un departamento solo tiene datos en 2023, o solo antes de 2023.
    # El modelo de Random Forest debería manejar esto, pero es una buena observación.
```

# Entrenamiento de los modelos
Se procederá a entrenar dos modelos RandomForestRegressor: uno para la predicción nacional y otro para la predicción a nivel departamental. Se utilizarán los datos de 2020 a 2022 para el entrenamiento y los datos de 2023 para la evaluación de ambos modelos.

```python
# Inicializar y entrenar el modelo RandomForestRegressor para la predicción NACIONAL
model_national = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

print("Entrenando el modelo RandomForestRegressor (Nacional)...")
model_national.fit(X_train_national, y_train_national)
print("¡Modelo Nacional entrenado exitosamente!")
```

```python
# Inicializar y entrenar el modelo RandomForestRegressor para la predicción DEPARTAMENTAL
model_departmental = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

print("\nEntrenando el modelo RandomForestRegressor (Departamental)...")
model_departmental.fit(X_train_departmental, y_train_departmental)
print("¡Modelo Departamental entrenado exitosamente!")
```

# Evaluación de los modelos
Una vez entrenados los modelos, se evaluará su rendimiento utilizando métricas comunes para problemas de regresión. Esto nos permitirá entender qué tan bien cada modelo predice los nacimientos en su respectivo conjunto de prueba (datos de 2023).

Las métricas a utilizar son:
- MAE (Mean Absolute Error - Error Absoluto Medio): Representa la magnitud promedio de los errores en un conjunto de predicciones, sin considerar su dirección. Es robusto a los valores atípicos.
- MSE (Mean Squared Error - Error Cuadrático Medio): Mide el promedio de los cuadrados de los errores. Penaliza más los errores grandes.
- RMSE (Root Mean Squared Error - Raíz del Error Cuadrático Medio): Es la raíz cuadrada del MSE. Tiene la misma unidad que la variable objetivo, lo que facilita su interpretación.
- R² (Coefficient of Determination - Coeficiente de Determinación): Indica la proporción de la varianza en la variable dependiente que es predecible a partir de las variables independientes. Un valor de 1.0 indica un ajuste perfecto.

```python
# Realizar predicciones en el conjunto de prueba (Nacional)
y_pred_national = model_national.predict(X_test_national)

# Calcular métricas de evaluación (Nacional)
mae_national = mean_absolute_error(y_test_national, y_pred_national)
mse_national = mean_squared_error(y_test_national, y_pred_national)
rmse_national = np.sqrt(mse_national)
r2_national = r2_score(y_test_national, y_pred_national)

print(f"\n--- Métricas de Evaluación del Modelo Nacional ---")
print(f"MAE (Error Absoluto Medio): {mae_national:.2f}")
print(f"MSE (Error Cuadrático Medio): {mse_national:.2f}")
print(f"RMSE (Raíz del Error Cuadrático Medio): {rmse_national:.2f}")
print(f"R² (Coeficiente de Determinación): {r2_national:.4f}")
```

# Interpretación de las métricas (Nacional):
- MAE de 4944.09 significa que, en promedio, las predicciones del modelo difieren de los valores reales en aproximadamente 4944.09 nacimientos a nivel nacional.
- RMSE de 5131.69 indica la magnitud típica de los errores de predicción en las unidades de nacimientos a nivel nacional.
- Un R² de -7.3574 sugiere que el -735.74% de la varianza en el número de nacimientos nacionales puede ser explicada por el modelo. Cuanto más cercano a 1, mejor es el ajuste del modelo a los datos.

```python
# Realizar predicciones en el conjunto de prueba (Departamental)
y_pred_departmental = model_departmental.predict(X_test_departmental)

# Calcular métricas de evaluación (Departamental)
mae_departmental = mean_absolute_error(y_test_departmental, y_pred_departmental)
mse_departmental = mean_squared_error(y_test_departmental, y_pred_departmental)
rmse_departmental = np.sqrt(mse_departmental)
r2_departmental = r2_score(y_test_departmental, y_pred_departmental)

print(f"\n--- Métricas de Evaluación del Modelo Departamental ---")
print(f"MAE (Error Absoluto Medio): {mae_departmental:.2f}")
print(f"MSE (Error Cuadrático Medio): {mse_departmental:.2f}")
print(f"RMSE (Raíz del Error Cuadrático Medio): {rmse_departmental:.2f}")
print(f"R² (Coeficiente de Determinación): {r2_departmental:.4f}")
```

# Interpretación de las métricas (Departamental):
- MAE de 158.28 significa que, en promedio, las predicciones del modelo difieren de los valores reales en aproximadamente 158.28 nacimientos a nivel departamental.
- RMSE de 226.30 indica la magnitud típica de los errores de predicción en las unidades de nacimientos a nivel departamental.
- Un R² de 0.9722 sugiere que el 97.22% de la varianza en el número de nacimientos departamentales puede ser explicada por el modelo. Cuanto más cercano a 1, mejor es el ajuste del modelo a los datos.

Un R² alto sugiere que el modelo es capaz de capturar las variaciones temporales y las diferencias entre departamentos. Es importante recordar que el error absoluto (MAE, RMSE) debe interpretarse en el contexto de la escala de nacimientos de cada departamento.
- Las visualizaciones son fundamentales para ambos modelos. Para el modelo nacional, nos permiten ver la alineación general de las predicciones. Para el modelo departamental, los gráficos por departamento y la distribución de residuos por departamento nos ayudan a identificar si el modelo se ajusta bien a las dinámicas específicas de cada región o si hay departamentos donde el rendimiento es subóptimo.

## Evaluación de la efectividad para predecir nacimientos en Colombia:

Ambos modelos son efectivos para capturar la tendencia general y la estacionalidad de los nacimientos en sus respectivos niveles. Esto es crucial para la planificación a corto y mediano plazo. Sin embargo, es importante considerar:

- Limitaciones de datos: Ambos modelos se entrenaron con solo 4 años de datos. Una serie histórica más extensa podría mejorar significativamente la capacidad de los modelos para identificar patrones a largo plazo y estacionalidades más complejas.
- Factores externos no considerados: Los modelos actuales solo utilizan el tiempo (año y mes) y el departamento como características. Factores socioeconómicos, políticos, de salud pública (ej. pandemias, programas de planificación familiar), migraciones, y eventos específicos (ej. desastres naturales) pueden influir en la natalidad y no están incluidos. La incorporación de estas variables externas podría aumentar la precisión.
- Heterogeneidad departamental (Modelo departamental): Aunque el enfoque global con one-hot encoding es un buen punto de partida, los departamentos pueden tener dinámicas de natalidad muy diferentes. Un modelo global podría tener dificultades para capturar todas estas particularidades si las interacciones no son lo suficientemente complejas.
- Tamaño del conjunto de prueba: Aunque la cantidad total de microdatos es grande, la serie temporal por mes (nacional) o por mes y departamento (departamental) para un año (12 muestras por serie) sigue siendo relativamente corta. Esto puede limitar la capacidad del modelo para aprender patrones muy complejos y evaluar su generalización a futuro.

## Posibles mejoras futuras:

- Ampliar series históricas: Obtener y utilizar datos de nacimientos de más años anteriores para ambos niveles de análisis.
- Incorporar Variables Externas: Incluir características como indicadores económicos (PIB, desempleo), políticas de salud, eventos demográficos (migración), y variables específicas de cada departamento (ej. tasa de pobreza departamental, acceso a servicios de salud, etc.).
- Modelos específicos por departamento: Si el rendimiento del modelo departamental global no es suficiente para ciertos departamentos clave, considerar entrenar modelos individuales para esas regiones con dinámicas muy particulares.
- Modelos jerárquicos o multi-serie: Explorar técnicas más avanzadas diseñadas para pronóstico de múltiples series temporales relacionadas, como modelos jerárquicos que modelan la relación entre el total nacional y los departamentos, o modelos de series temporales multivariadas.
- Modelos de series temporales avanzados: Explorar modelos como ARIMA, SARIMA, Prophet o redes neuronales recurrentes (RNNs) que están diseñados específicamente para datos de series temporales y pueden capturar dependencias temporales más complejas.
- Optimización de hiperparámetros: Realizar una búsqueda de hiperparámetros (ej. GridSearchCV, RandomizedSearchCV) para los RandomForestRegressor o cualquier otro modelo elegido para encontrar la configuración óptima.
